```python
# improved version of the code:


def calculator_bmi(weight_kg, height_m):
    return weight_kg / (height_m ** 2)

def convert_to_metric(weight_lb, height_in):
    weight_kg = weight_lb * 0.453592
    height_m = height_in * 0.0254
    return weight_kg, height_m

def get_bmi_category(bmi):
    if bmi < 18.5:
        return "underweight", "https://www.bmi-calculator.net/treatment/underweight-treatment.php"
    elif 18.5 <= bmi < 24.9:
        return "normal weight", "https://www.bmi-calculator.net/treatment/normal-weight-treatment.php"
    elif 25 <= bmi < 29.9:
        return "overweight", "https://www.bmi-calculator.net/treatment/overweight-treatment.php"
    elif 30 <= bmi < 34.9:
        return "obesity (Class I)", "https://www.bmi-calculator.net/treatment/obesity-treatment.php"
    elif 35 <= bmi < 39.9:
        return "obesity (Class II)", "https://www.bmi-calculator.net/treatment/obesity-treatment.php"
    else:
        return "morbid obesity (Class III)", "https://www.bmi-calculator.net/treatment/obesity-treatment.php"

name = input("Enter your name: ")
weight_lb = float(input("Enter your weight in pounds: "))
height_in = float(input("Enter your height in inches: "))

# Convert to metric units
weight_kg, height_m = convert_to_metric(weight_lb, height_in)

# Calculate BMI
bmi = calculator_bmi(weight_kg, height_m)

# Get BMI category and treatment info
category, treatment_url = get_bmi_category(bmi)

print(f"{name}, your BMI is {bmi:.2f}. You are {category}.")
print(f"Treatment info:  {treatment_url}")
```
